```python
import ipywidgets as widgets
from IPython.display import display, HTML

# Title
display(HTML("<h2 style='color:#2C3E50; text-align:center;'>üßÆ Python Arithmetic Calculator</h2>"))

# Input fields
num1 = widgets.FloatText(
    description='Please enter 1st number:',
    style={'description_width': '200px'},
    layout=widgets.Layout(width='320px')
)

num2 = widgets.FloatText(
    description='Please enter 2nd number:',
    style={'description_width': '200px'},
    layout=widgets.Layout(width='320px')
)

# Calculate button
button = widgets.Button(
    description="Calculate",
    button_style='info',
    tooltip='Click to calculate results',
    layout=widgets.Layout(width='150px', margin='10px 0 10px 0')
)

# Output area
output = widgets.Output(layout={'border': '1px solid #ccc', 'padding': '10px', 'width': '400px'})

# Logic for calculations
def calculate(_):
    with output:
        output.clear_output()
        n1, n2 = num1.value, num2.value
        print(f"Sum: {n1 + n2}")
        print(f"Difference: {n1 - n2}")
        print(f"Product: {n1 * n2}")
        if n2 != 0:
            print(f"Quotient: {n1 / n2}")
            print(f"Remainder: {n1 % n2}")
        else:
            print("Division or modulus by zero is not allowed.")
        try:
            if abs(n2) > 100:
                print("Power: Too large to calculate safely ‚ö†Ô∏è")
            else:
                print(f"Power: {n1 ** n2}")
        except OverflowError:
            print("Power: Calculation overflowed (result too large)")

# Link button with function
button.on_click(calculate)

# Arrange layout vertically & center align
ui = widgets.VBox([
    widgets.HBox([num1]),
    widgets.HBox([num2]),
    widgets.HBox([button]),
    output
])

display(widgets.Box([ui], layout=widgets.Layout(
    display='flex',
    justify_content='center',
    align_items='center'
)))
```
