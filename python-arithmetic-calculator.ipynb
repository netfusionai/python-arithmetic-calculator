```python
import ipywidgets as widgets
from IPython.display import display

# Create input boxes
num1 = widgets.FloatText(description='Number 1:', value=0)
num2 = widgets.FloatText(description='Number 2:', value=0)

# Output area
output = widgets.Output()

# Function to perform calculations
def calculate(_):
    with output:
        output.clear_output()
        n1, n2 = num1.value, num2.value
        print(f"Sum: {n1 + n2}")
        print(f"Difference: {n1 - n2}")
        print(f"Product: {n1 * n2}")
        if n2 != 0:
            print(f"Quotient: {n1 / n2}")
            print(f"Remainder: {n1 % n2}")
        else:
            print("Division or modulus by zero is not allowed.")

        # Safe exponentiation
        try:
            if abs(n2) > 100:
                print("Power: Too large to calculate safely ⚠️")
            else:
                print(f"Power: {n1 ** n2}")
        except OverflowError:
            print("Power: Calculation overflowed (result too large)")

# Button to trigger
button = widgets.Button(description="Calculate")
button.on_click(calculate)

# Display UI
display(num1, num2, button, output)
```
