```python
## lets create a simple calculator by taking two numbers as input from user to do basic maths

num1=float(input("Enter first number: ")) # converting str into float 
num2=float(input("Enter second number: ")) # converting str into float

sum=num1+num2
product=num1*num2
difference=num1-num2
quotient=num1/num2
modulus=num1%num2
exponentiation=num1**num2

print("your first entered number is: ", num1 ,"and second number is: ", num2)
print("Sum is: ",sum)
print("Difference between two numbers is: ",difference)
print("Product is: ",product)
print("Quotient is: ",quotient)
print("The remainder between both numbers is: ",modulus)
print("first number to the power of second number is: ",exponentiation)
```
