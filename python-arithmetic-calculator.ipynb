```python
import ipywidgets as widgets
from IPython.display import display

# Create input boxes
num1 = widgets.FloatText(description='Number 1:', value=0)
num2 = widgets.FloatText(description='Number 2:', value=0)

# Create output area
output = widgets.Output()

# Define function
def calculate(_):
    with output:
        output.clear_output()
        n1, n2 = num1.value, num2.value
        print(f"Sum: {n1 + n2}")
        print(f"Difference: {n1 - n2}")
        print(f"Product: {n1 * n2}")
        if n2 != 0:
            print(f"Quotient: {n1 / n2}")
            print(f"Remainder: {n1 % n2}")
        else:
            print("Division or modulus by zero is not allowed.")
        print(f"Power: {n1 ** n2}")

# Button to trigger calculation
button = widgets.Button(description="Calculate")
button.on_click(calculate)

# Display everything
display(num1, num2, button, output)
```
